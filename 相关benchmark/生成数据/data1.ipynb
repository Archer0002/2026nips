```python
import numpy as np

def generate_data(n=1000, seed=42):
    np.random.seed(seed)
    
    # x_i ~ N(0, I_{10x10})
    X = np.random.multivariate_normal(mean=np.zeros(10), cov=np.eye(10), size=n)
    
    # t_i | x_i ~ Bern(1 / (1 + exp(-x_{i,3})))
    # 注意: Python索引从0开始，x_{i,3} 对应 X[:, 2]
    propensity = 1.0 / (1.0 + np.exp(-X[:, 2]))
    t = np.random.binomial(1, propensity)
    
    # epsilon_i | x_i, t_i ~ N(0, 1)
    epsilon = np.random.normal(0, 1, size=n)
    
    # tau | x_i = 1 + 2|x_{i,4}| + x_{i,10}^2
    # x_{i,4} -> X[:, 3], x_{i,10} -> X[:, 9]
    tau = 1 + 2 * np.abs(X[:, 3]) + X[:, 9] ** 2
    
    # y_i | x_i, tau_i, epsilon_i, t_i = t_i * tau_i + epsilon_i
    #   + 5 * (2 + 0.5*sin(pi*x_{i,1}) - 0.5*x_{i,2} + 0.75*x_{i,3}*x_{i,9})
    # x_{i,1} -> X[:, 0], x_{i,2} -> X[:, 1], x_{i,3} -> X[:, 2], x_{i,9} -> X[:, 8]
    baseline = 5 * (2 + 0.5 * np.sin(np.pi * X[:, 0]) - 0.5 * X[:, 1] + 0.75 * X[:, 2] * X[:, 8])
    y = t * tau + epsilon + baseline
    
    return X, t, tau, y, epsilon

if __name__ == "__main__":
    X, t, tau, y, epsilon = generate_data(n=5000)
    
    print(f"样本数: {len(y)}")
    print(f"X shape: {X.shape}")
    print(f"处理组比例: {t.mean():.3f}")
    print(f"tau 均值: {tau.mean():.3f}, 标准差: {tau.std():.3f}")
    print(f"y 均值: {y.mean():.3f}, 标准差: {y.std():.3f}")
```
